```python
# TODO: uncomment the packages you used, please do not import additional non-native packages
# you may change the imports to the following format: from [package] import [class, method, etc.]

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture


import scipy
import seaborn as sns
```

K-Means Clustering:


```python
# Load the dataset
data = pd.read_csv('data/dataset1_noClusters2.csv')

# Remove the labels
features = data.iloc[:, :-1]

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(features)

# Visualize the clusters
plt.scatter(features.iloc[:, 0], features.iloc[:, 1], c=labels, cmap='viridis')
plt.title('K-Means Clustering (Dataset 1)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

Hierarchical Clustering (Average Linkage):


```python

# Load the dataset
data = pd.read_csv('data/dataset2_noClusters2.csv')

# Remove the labels
features = data.iloc[:, :-1]

# Apply Hierarchical clustering (Average Linkage)
hc = AgglomerativeClustering(n_clusters=2, linkage='average')
labels = hc.fit_predict(features)

# Visualize the clusters
plt.scatter(features.iloc[:, 0], features.iloc[:, 1], c=labels, cmap='viridis')
plt.title('Hierarchical Clustering (Dataset 1)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

DBSCAN (Density-Based Spatial Clustering of Applications with Noise):


```python

# Load the dataset
data = pd.read_csv('data/dataset2_noClusters2.csv')

# Remove the labels
features = data.iloc[:, :-1]

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(scaled_features)

# Visualize the clusters
plt.scatter(features.iloc[:, 0], features.iloc[:, 1], c=labels, cmap='viridis')
plt.title('DBSCAN Clustering (Dataset 1)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
data = pd.read_csv('data/dataset4_noClusters7.csv')

# Remove the labels
features = data.iloc[:, :-1]

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply Expectation Maximization
em = GaussianMixture(n_components=2, random_state=42)
labels = em.fit_predict(scaled_features)

# Visualize the clusters
plt.scatter(features.iloc[:, 0], features.iloc[:, 1], c=labels, cmap='viridis')
plt.title('Expectation Maximization (Dataset 1)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
